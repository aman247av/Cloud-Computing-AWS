### Service Models

a specific framework or approach that defines the type of services and level of control offered by a cloud provider to its customers. It outlines the responsibilities and functionalities shared between the cloud provider and the cloud user (customer) Eg: Amazon Web Services (AWS)

- **Infrastructure as a Service (IaaS):** In the IaaS service model, cloud providers offer virtualized computing resources over the internet. These resources include virtual machines, storage, and networking components. 

- **Platform as a Service (PaaS):** PaaS provides a platform that includes the infrastructure, runtime environment, and development tools for building, deploying, and managing applications. Eg: Heroku

- **Software as a Service (SaaS):** SaaS is a service model where cloud providers deliver complete software applications over the internet on a subscription basis. Eg:  Google Workspace




### Deployment Models

refer to various ways or configurations in which cloud computing resources and services can be organized and delivered to users. These models define the physical and virtual infrastructure setup and how the cloud environment is structured.

-  **Public Cloud:** Public clouds are typically multi-tenant environments, meaning multiple customers share the same infrastructure and resources. 

- **Private Cloud:** In a private cloud deployment, the cloud infrastructure is dedicated to a single organization or entity. It can be hosted on-premises within an organization's data center or provided by a third-party provider.

- **Hybrid Cloud:** A hybrid cloud deployment combines elements of both public and private clouds, allowing data and applications to be shared between them.

Virtualization is the ability to run multiple

#### Benefits of Virtualization
 - Sharing of resources helps cost reduction - Isolation: Virtual machines are isolated from each other as if they are physically separated - Encapsulation: Virtual machines encapsulate a complete computing environment - Hardware Independence: Virtual machines run independently of underlying hardware - Portability: Virtual machines can be migrated between different hosts.

**Virtual Machines (VMs):** Virtualization enables the creation of virtual machines, which are software-based representations of physical computers. Each VM runs its own operating system and applications independently of the physical hardware. 

**Hypervisor:** The hypervisor, also known as a virtual machine monitor (VMM), is a critical component of virtualization. It manages the virtualization process by allocating and managing resources, such as CPU, memory, and storage, to each VM. There are two types of hypervisors: 
- Type 1 (bare-metal) hypervisors run directly on physical hardware, while 
- Type 2 (hosted) hypervisors run on top of an existing operating system.


#### 1. Emulation or simulation

" allowing an unmodified "guest" OS for a
completely different CPU to be run. "

The term "guest" OS refers to the operating system that was originally designed to run on a specific CPU architecture. It's called "unmodified" because you don't need to change the guest OS to make it work on the emulator, which simulates the behavior of a different CPU architecture. It acts like a virtual version of the target CPU. When you run an emulator on your computer or device, it essentially pretends to be the CPU that the guest OS was built for.

#### 2. Native and Full Virtualization

" the virtual machine simulates enough
hardware to allow an unmodified "guest" OS
(one designed for the same CPU) to be run in
isolation. "

native virtualization and full virtualization allow unmodified guest OSes designed for the same CPU architecture to run in isolated virtual machines. 


#### 3. Hardware enabled virtualization

" the virtual machine has its own hardware and
allows a guest OS to be run in isolation."

hardware-enabled virtualization optimizes virtualization performance and security by using CPU hardware extensions to allow VMs to directly access physical hardware resources while maintaining isolation and control through the hypervisor.
- A hypervisor, or virtualization layer, initializes and manages VMs directly on the physical hardware. 
- When a VM is running, the hypervisor takes advantage of CPU hardware extensions (VT-x or AMD-V) to allow certain guest OS instructions to be executed directly on the physical CPU without translation or emulation.
- The hypervisor manages CPU time, memory allocation, and I/O operations, ensuring fair resource access among VMs.


#### 4. Paravirtualization

"the virtual machine does not necessarily
simulate hardware, but instead (or in addition)
offers a special API that can only be used by
modifying the "guest" OS."

 Unlike full virtualization, where VMs mimic physical hardware, paravirtualization introduces a special API within the VM. This API is designed to provide a more efficient and direct way for the VM to communicate with the hypervisor and access certain hardware resources.

 unlike traditional full virtualization where virtual machines (VMs) mimic physical hardware to a great extent, in paravirtualization, the VM doesn't try to replicate all aspects of physical hardware. Instead, it focuses on providing a specialized software interface (API) that can only be utilized by making modifications to the "guest" operating system (OS) to communicate more directly with the hypervisor<br>KVM,
 

#### 5. Operating system-level virtualization

"virtualizing a physical server at the operating
system level, enabling multiple isolated and
secure virtualized servers to run on a single
physical server."

Host Operating System: The physical server, also known as the host, runs a complete operating system. This host OS is responsible for managing the server's hardware resources, including CPU, memory, storage, and networking.

Containerization Software: A containerization software platform, such as Docker or Kubernetes, is installed on the host OS. This software provides the tools and infrastructure needed to create, manage, and run containers.

Isolation: Containers are isolated from each other and from the host OS. This isolation ensures that processes running within one container cannot interfere with processes in other containers or the host system. Isolation is achieved through various Linux kernel features, such as namespaces and cgroups.

Resource Allocation: The containerization software allocates a portion of the host's resources (CPU, memory, storage, etc.) to each container based on predefined resource limits and requirements. 
Eg: Parallels Workstation, Solaris Containers

## Load Balancing

Cloud computing resources can be scaled up
on demand to meet the performance
requirements of applications.

• Load balancing distributes workloads across
multiple servers to meet the application
workloads.

The goals of load balancing techniques include:
• Achieve maximum utilization of resources
• Minimizing the response times
• Maximizing throughput

In a Cluster-Based Web System:

- Virtual IP (VIP) is used.
- A Load Balancer (Web Switch) is in front of server nodes.
- Clients send requests to the VIP.
- Load balancer distributes traffic to server nodes.
- Load balancing is based on various algorithms.
- Server nodes' IPs are hidden from clients.
- Clients interact with the VIP, not individual server nodes.


In a Distributed Web System:

- Each web server node has its own unique IP address visible to clients.
- Clients send requests directly to the IP addresses of the server nodes.
- A Layer 3 router routes incoming requests based on destination IP addresses but doesn't perform load balancing.
- Clients communicate directly with individual server nodes, knowing which node they are interacting with.

## Network Function Virtualization

• Network Function Virtualization (NFV) is a technology that leverages
virtualization to consolidate the heterogeneous network devices onto industry
standard high volume servers, switches and storage.

 Key elements of the NFV architecture are

• Virtualized Network Function (VNF): VNF is
a software implementation of a network
function which is capable of running over
the NFV Infrastructure (NFVI).

• NFV Infrastructure (NFVI): NFVI includes
compute, network and storage resources that
are virtualized.

• NFV Management and Orchestration: NFV
Management and Orchestration focuses on all
virtualization-specific management tasks and
covers the orchestration and lifecycle
management of physical and/or software
resources that support the infrastructure
virtualization, and the lifecycle management of
VNFs.

## MapReduce
MapReduce is a parallel data processing
model for processing and analysis of massive
scale data.

MapReduce phases:

• Map Phase: In the Map phase, data is read from a
distributed file system, partitioned among a set of
computing nodes in the cluster, and sent to the nodes
as a set of key-value pairs.

• The Map tasks process the input records
independently of each other and produce
intermediate results as key-value pairs.

• The intermediate results are stored on the local disk
of the node running the Map task.

• Reduce Phase: When all the Map tasks are
completed, the Reduce phase begins in which the
intermediate data with the same key is aggregated.
